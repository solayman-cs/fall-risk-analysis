```python
import pandas as pd


file_path = 'hospital-fall-data-2012-2017.xlsx'
xls = pd.ExcelFile(file_path)


merged_data = pd.DataFrame()


for sheet_name in xls.sheet_names:
   
    current_data = pd.read_excel(file_path, sheet_name)
    current_data['Year'] = int(sheet_name)  

    merged_data = pd.concat([merged_data, current_data], ignore_index=True)


print(merged_data)

# Save the merged dataset to a new Excel file 
merged_data.to_excel('merged_dataset_fall.xlsx', index=False)
```
